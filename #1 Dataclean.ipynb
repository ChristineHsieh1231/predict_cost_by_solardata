## 匯入另一筆史丹佛的DeepSolar研究資料

```python

deepsolar = pd.read_csv("/Users/christine_sieh/workspace/Electric price_caculator/#1 dataclean/deepsolar_tract.csv")
deepsolar.info()
```

把 Google太陽能屋頂的資料與 Stanford Deep Solar 兩個資料用`county`, `state`兩個欄位做雙重合併<br>因為發現不同州別也會有相同的鄉鎮名稱，所以必須用雙重條件合併欄位


```python
deepsolar
```

```python
deepsolar.info(verbose=True)
```

```python
solar_roof = sunroof.merge(deepsolar, how='inner', left_on=['region_name', 'state_name'], right_on=['county', 'state'])
solar_roof
```

```python
#check
solar_roof[['region_name','state_name','county','state']]
```

找出依變數(dependent variable)：y

`electricity_price_overall`(cents/kWh)$*$ `electricity_consume_total` (kWh)$*12$ <br> `Electricity_cost_total` 年度電費總花費(cents)

找出年度電價總花費額

```python
solar_roof['electricity_cost_total']= (solar_roof['electricity_price_overall']*solar_roof['electricity_consume_total']*12)
```

1. `electricity_consume_residential`
2. `electricity_consume_commercial`
3. `electricity_consume_industrial`
4. `electricity_consume_total`<br>

太陽能預計發電量（google）的欄位是以年為單位 但是史丹佛 Deep solar 這四個欄位資料是用月份作為單位<br>
這4欄位的單位是用年去算出的平均數，把史丹佛 Deep solar 四個欄位的數字同乘12

方法參考pandas官網[05_add_columns](https://pandas.pydata.org/pandas-docs/stable/getting_started/intro_tutorials/05_add_columns.html)
